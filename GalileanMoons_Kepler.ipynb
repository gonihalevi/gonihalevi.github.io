```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams['font.family'] ='serif'
mpl.rcParams['font.size'] = 16
```

```python
# data from https://nssdc.gsfc.nasa.gov/planetary/factsheet/galileanfact_table.html
# in order of closest to furthest: Io, Europa, Ganymede, Callisto
a = 1e6*np.array([422,671,1070,1883]) #semimajor axis in 10^3 km converted into m since it is what most students did
P = 24*3600*np.array([1.769,3.551,7.155,16.689]) #orbital period in days converted into seconds
```

```python
x, y = np.log10(a),np.log10(P)

line = np.poly1d(np.polyfit(x, y, 1))
m,b = line #getting slope and y-intercept of best-fit line
print(f'slope = {m}')
print(f'y-intercept = {b}')

x_line = np.linspace(8.4,9.4,4)
plt.figure(figsize=(7,4))
plt.xlabel('log(a [cm])')
plt.ylabel('log(P [sec])')
plt.grid()
plt.plot(x_line, line(x_line),linestyle='dashed',lw=3,color='k',label=f'log(P)={np.round(m,2)}log(a)+({np.round(b,2)})')
plt.plot(x,y,color='purple',label='data',marker='o',ls='',markersize=10)
plt.xlim((8.4,9.4))
plt.ylim((4.6,6.4))
plt.legend()
plt.savefig('/Users/ghalevi/Dropbox/IIT/360/Spring2025/GalileanMoons_Kepler.png',bbox_inches='tight',dpi=100)
```

$$P^2 = \frac{4\pi^2}{GM_\mathrm{Jup}} a^3$$
$$\log{(P^2)} = \log{\left(\frac{4\pi^2}{GM_\mathrm{Jup}}\right)} + \log{(a^3)}$$
$$2 \log{(P)} = \log{\left(\frac{4\pi^2}{GM_\mathrm{Jup}}\right)} + 3\log{(a)}$$
$$\log{(P)} = \frac{1}{2} \log{\left(\frac{4\pi^2}{GM_\mathrm{Jup}}\right)} + \frac{3}{2}\log{(a)}$$
Comparing this to the form for a line, $y = mx+b$, we see that the slope, $m$, is $3/2$ (as we found for our best-fit) and the y-intercept, $b$, is $\frac{1}{2} \log{\left(\frac{4\pi^2}{GM_\mathrm{Jup}}\right)}$.
$$ 2b = \log{(4\pi^2)}-\log{(GM_\mathrm{Jup})}$$
$$ \log{(GM_\mathrm{Jup})} = \log{(4\pi^2)}-2 b$$
$$ GM_\mathrm{Jup} = 10^{\log{(4\pi^2)}-2 b}$$
$$ M_\mathrm{Jup} = 10^{\log{(4\pi^2)}-2 b}/G$$

```python
logGM = np.log10(4*np.pi**2)-2*b
```

```python
GM = 10**logGM
```

```python
G = 6.67e-11 #in SI
M = GM/G
```
